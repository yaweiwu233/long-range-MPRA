```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, linregress
```

```python
## Set figure paramaters
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial' 
plt.rcParams['svg.fonttype'] = 'none'
```

```python
## Set parameters
data_folder = "../data/individual_validation/"
figure_folder = "../figures/"
primers = ["hPRT", "mScarlet"]
distances = ["0kb", "10kb"]
promoters = ["HBG", "GAPDH", "HBE"]
```

```python
## Read qRT-PCR CT values for 7 enhancer + no-enhancer control
qpcr_data = pd.read_csv(data_folder+"HBG_qPCR.txt", sep="\t", header=None)
qpcr_data.columns = ["sample", "CT", "primer"]
```

```python
## Take average across technique replciates
averaged_data = qpcr_data.groupby(["sample", "primer"])["CT"].mean().reset_index()
averaged_data_long = averaged_data.pivot(index="sample", columns="primer", values="CT").reset_index()
```

```python
## Normalize by housekeeping gene
averaged_data_long["delta_CT"] = averaged_data_long["mScarlet"] - averaged_data_long["hPRT"]
```

```python
## Normalize by basal promoter
ctrl_0kb = averaged_data_long.loc[(averaged_data_long["sample"].str.contains("-0kb")) & (averaged_data_long["sample"].str.contains("ctrl")), "delta_CT"].mean()
ctrl_10kb = averaged_data_long.loc[(averaged_data_long["sample"].str.contains("-10kb")) & (averaged_data_long["sample"].str.contains("ctrl")), "delta_CT"].mean()
averaged_data_long.loc[averaged_data_long["sample"].str.contains("-0kb"), "ddCT"] = averaged_data_long.loc[averaged_data_long["sample"].str.contains("-0kb"), "delta_CT"]-ctrl_0kb
averaged_data_long.loc[averaged_data_long["sample"].str.contains("-10kb"), "ddCT"] = averaged_data_long.loc[averaged_data_long["sample"].str.contains("-10kb"), "delta_CT"]-ctrl_10kb
## Calculate foldchange
averaged_data_long["foldchange"] = 2**(averaged_data_long["ddCT"]*(-1))
## Parse variables
averaged_data_long["distance"] = averaged_data_long["sample"].str.split("-").str[2].str.split("+").str[0]
averaged_data_long["enhancer"] = averaged_data_long["sample"].str.split("+").str[1].str.split("-").str[0]
averaged_data_long["name"] = averaged_data_long["sample"].str.replace("-rep1", "").str.replace("-rep2", "")
```

```python
## Read long-range MPRA data
mpra = pd.read_csv(data_folder+"long_range_MPRA_data.txt", sep="\t")
mpra["enhancer"] = mpra["Name"].str.split("-enh").str[1]
mpra_long = mpra.melt(id_vars = "enhancer", value_vars = ["HBG_0kb", "HBG_10kb"], var_name = "distance", value_name="mpra")
mpra_long["distance"] = mpra_long["distance"].str.strip("HBG_")
## Merge two datasets
merged = pd.merge(averaged_data_long, mpra_long, on=["enhancer", "distance"])
```

```python
## Set color scheme
distance_2LP = {
    '0kb': "plum",  # Lightest shade
    '10kb': "purple",  # Distinct shade for frequent use
}
```

```python
## Plot correlation between long-range MPRA and RT-qPCR
plt.figure(figsize=(2.2,2.2))
sns.lineplot(merged, x="mpra", y="foldchange", hue="distance", linestyle="", err_style="bars", marker="o", palette=distance_2LP, mec='none', markersize=4.5)
plt.legend(fontsize=8)
plt.xlabel("Enhancer activity\n(long-range MPRA)", fontsize=8)
plt.ylabel("Enhancer activity\n(RT-qPCR)", fontsize=8)
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
r1, p1 = pearsonr(merged.loc[merged["distance"] == "0kb", "mpra"], merged.loc[merged["distance"] == "0kb", "foldchange"])
r2, p2 = pearsonr(merged.loc[merged["distance"] == "10kb", "mpra"], merged.loc[merged["distance"] == "10kb", "foldchange"])
plt.text(4.5, 1.2, f"R={r1:.2f}", color="plum", fontsize=8)
plt.text(4.5, 0.2, f"R={r2:.2f}", color="purple", fontsize=8)
## Add fit line
for d, color in distance_2LP.items():
    subset = merged[merged["distance"] == d]
    sns.regplot(data=subset, x="mpra", y="foldchange", scatter=False,
                line_kws={"color": color, "alpha": 0.2, "linewidth": 1})
plt.legend(fontsize=8, frameon=False)
plt.xlabel("Enhancer activity\n(long-range MPRA)", fontsize=8)
plt.ylabel("Enhancer activity\n(RT-qPCR)", fontsize=8)
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
plt.savefig(figure_folder+"individual_validation_qpcr_vs_mpra.svg", dpi=300)
```

```python
## Read fluorescence data for 20 samples with HBG promoter
flow_data = pd.read_csv(data_folder+"PE_mean_flow_HBG_all_enhancers.txt", sep="\t", header=None)
flow_data.columns = ["name", "MFI"]
flow_data["enhancer"] = flow_data["name"].str.split("+").str[1]
flow_data["promoter"] = flow_data["name"].str.split("-").str[1]
flow_data["distance"] = flow_data["name"].str.split("-").str[2].str.split("+").str[0]
```

```python
## Normalize by no-enhancer control
for x in promoters:
    for y in distances:
        baseline = flow_data.loc[(flow_data["promoter"] == x) & (flow_data["distance"] == y) & (flow_data["enhancer"] == "ctrl"), "MFI"].mean()
        flow_data.loc[(flow_data["promoter"] == x) & (flow_data["distance"] == y), "norm_MFI"] = flow_data.loc[(flow_data["promoter"] == x) & (flow_data["distance"] == y), "MFI"]/baseline
```

```python
## Merge two datasets
merged = pd.merge(flow_data, mpra_long, on=["enhancer", "distance"])
```

```python
## Plot correlation between long-range MPRA and flow cytometry
plt.figure(figsize=(2.2,2.2))
sns.lineplot(merged, x="mpra", y="norm_MFI", hue="distance", linestyle="", err_style="bars", marker="o", palette=distance_2LP, markersize = 4, mec='none')

r1, p1 = pearsonr(merged.loc[merged["distance"] == "0kb", "mpra"], merged.loc[merged["distance"] == "0kb", "norm_MFI"])
r2, p2 = pearsonr(merged.loc[merged["distance"] == "10kb", "mpra"], merged.loc[merged["distance"] == "10kb", "norm_MFI"])
plt.text(4.5, 1.1, f"R={r1:.2f}", color="plum", fontsize=8)
plt.text(4.5, 0.5, f"R={r2:.2f}", color="purple", fontsize=8)
for d, color in distance_2LP.items():
    subset = merged[merged["distance"] == d]
    sns.regplot(data=subset, x="mpra", y="norm_MFI", scatter=False,
                line_kws={"color": color, "alpha": 0.2, "linewidth": 0.75})
plt.legend(fontsize=8, frameon=False)
plt.xlabel("Enhancer activity\n(long-range MPRA)", fontsize=8)
plt.ylabel("Enhancer activity\n(flow cytometry)", fontsize=8)
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
plt.savefig(figure_folder+"individual_validation_flow_vs_mpra.svg", dpi=300)
```
